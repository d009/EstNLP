### Süntaktiline analüüs

Süntaktilise analüüsi (sünonüümina kasutatakse sõna "parsima") ülesandeks on lause struktuuri ehk sõnadevaheliste seoste leidmine. Kui me teostame tekstil morfoloogilise analüüsi ja leiame sõnaliigid, siis saame üldjoontes teada, milliseid tegevusi selle lausega kirjeldatakse (tegusõnad), millistest olenditest/objektidest/nähtustest jne seal räägitakse (nimisõnad), milliseid omadusi seal mainitakse (omadussõnad) jne. Sõnaliigid ei anna aga meile infot sõnadevahelistest seostest. Näiteks, kui teame, et lauses esinevad nimisõnad "ema" ja "isa" ning tegusõna "hirmutama", siis ei ole meil siiski teadmist selle kohta, kes keda ehmatab. Kui teame aga, et "ema" on lauses alus (subjekt), "isa" on sihitis (objekt) ning "hirmutama" on selle lause öeldis, siis on selge, et ema on hirmutajaks ja isa selleks, keda hirmutatakse. Seda teadmist saame esitada puu kujul:

Eesti keele süntaktiliseks märgendamiseks on kaks enamkasutatavat töövahendit statistiline [MaltParser](http://www.maltparser.org/) ja reeglipõhine kitsenduste grammatika formalismil baseeruv [VISLCG3](https://github.com/EstSyntax/EstCG) analüsaator. Mõlemat on võimalik kasutada läbi EstNLTK (vt ka [juhend](https://github.com/estnltk/estnltk/blob/version_1.6/tutorials/syntax/syntax.ipynb)) ning mõlema väljundit aitab vajadusel mõista praktikumikaustas olev *syntaksiliides_ee.pdf*

### VISLCG3

VISLCG3 parseri kasutamiseks on vaja VISLCG3 esmalt installida oma arvutisse (vt juhendit [siit](http://visl.sdu.dk/cg3/single/#installation)). Selle kasutamiseks läbi EstNLTK tuleb VislTagger isendit luues lipuga `vislcg_cmd` ette anda vislcg3 asukoha oma arvutis. Kuna aga edaspidi on praktikumis aga vajalik, et kausta tee oleks süsteemi-muutujasse lisatud, siis seda ka teeme ja parseri isendi loomisel pole tarvis vislcg3-e asukohta ette anda.

```python
from estnltk.taggers import VislTagger
from estnltk import Text
```

```python
visl_tagger = VislTagger()  

visl_tagger.input_layers
```

Nagu näha, nõuab VISLCG3, et Text-objektil oleks kiht `morph_extended`. Lisame `morph_extended` kihi ja vaatame korraks, mis moel see tavapärasest morfoloogilise analüüsisi kihist erineb.

```python
text = Text('Poe ees õlut joonud mehed häirisid kohalikke, aga ükskõikne politsei ei teinud midagi.')
```

```python
text.tag_layer(['morph_extended'])
display(text['morph_extended'])
```

Näeme, et võrreldes 'morph_analysis' kihiga, sisaldab 'morph_extended' kihis atribuut *form* detailsemat infot ja juures on ka mitu uut atribuuti (nt *punctuation_type* ja *letter_case*) 

Märgendame teksti nüüd VISLCG3-ga:

```python
visl_tagger.tag(text)
text.layers
```

Kui tekst on analüüsitud VISLCG3 parseriga, lisandub tekstile kiht `visl`

```python
text['visl']
```

Atribuut `deprel` märgib parseri lisatud süntaktilist märgendit, nt @SUBJ väljendab (osa)lause subjekti @OBJ objekti, @FMV finiitset põhiverbi (öeldist). Märgendite tähendused leiab failist *syntaksiliides_ee.pdf*.

Atribuut `head` viitab lauseliikme ülemuse *id*-le. Näiteks allub subjekt "mehed" lauseliikmele, mille id on 6, ehk öeldisele "häirisid". Viimane omakorda on lause **juur**, st ei allu ühelegi teisele lauseliikmele, sellele viitab *head*-atribuudi väärtus 0.

Süntaksi märgendus võib jääda ka mitmeseks, sellegipoolest määratakse igale sõnale vaid üks ülemus. Nt järgnevas lauses saavad nimisõnad "tallinlased" ja "tartlased" nii subjekti kui ka öeldistäite märgendused, samuti on mitmene sõna "teine":

```python
text = Text('Ühel küljel on tallinlased ja teisel tartlased').tag_layer(['morph_extended'])
visl_tagger.tag(text)
text.visl
```

#### Puustruktuuri uurimine
Süntaksis ei huvitu me enamasti vaid ühest sõnast korraga, vaid ka sõnadevahelistest seostest. Nagu võib aimata varem läbi käinud mõistetest *juur* ja *ülemus*, moodustavad sõnad ja nendevahelised suhted (süntaksi)puu. Selleks, et ülemustele-alluvatele lihtsamini ligi pääseda ja meid huvitavaid konstruktsioone lihtsamini läbi käia, saab märgenduskihile lisada `SyntaxDependencyRetaggeri` abil atribuudid `parent_span` ja `children` (viitavad ülemuste-alluvate Span'idele). Vaatame näidet kahelauselise teksti puhul:

```python
text = Text("Pisike poiss kimbutas kirpe. Need kirbud ei julgenud kiusata karusid.")
text.tag_layer(['sentences', 'morph_extended'])
visl_tagger.tag(text)
text.visl
```

```python
from estnltk.taggers import SyntaxDependencyRetagger

SyntaxDependencyRetagger('visl').retag(text)

text.visl
```

Itereerime üle lausete finiitverbide alluvate ja väljastame nende ülemuse ja süntaktilise märgendi:

```python
for word in text.visl:
    if word.deprel[0] == '@FMV':
        for child in word.annotations[0].children:
            print(word.text, '-->', child.annotations[0].deprel, child.annotations[0].text)
        
```

Pane tähele, et alluvaid võib ühel lauseliikmel olla mitu, ülemaid aga üks. Pärime esimese sõna ülemuse ning suhte(d), millega see ülema külge seotud on :

```python
yhel = text.visl[0]
parent = yhel.annotations[0].parent_span.text
rel = yhel.deprel
print(yhel.text, rel, parent)
```

### Lisamaterjal: Univesal Dependencies ja CoNLL-U formaat

[Universal Dependencies](https://universaldependencies.org/#language-) (UD) on projekt, mille eesmärk on luua eri keeltele sobiv morfoloogiline ja sõltuvussüntaktiline märgendussüsteem ning seda järgivad puudepangad, mis võimaldaksid nt mitmekeelse parseri arendamise ja keeltevahelise masinõppe. Seega püüab UD universaalsete kategooriate abil kindlustada keelest sõltumatult järjepidevat sarnaste konstruktsioonide kirjeldamist, samas lubades vajadusel teatavaid keelespetsiifilisi täiendusi. UD süntaksimärgendamise põhimõtetest saab rohkem lugeda [siit](https://universaldependencies.org/u/overview/syntax.html).

UD peab olema rahuldav paljude keelte (tüpoloogia) kirjeldamiseks, samas lihtsasti mõistetav nii inimesele (sh mitte-lingvistile) kui ka sobiv arvutiga parsimiseks ja töötlemiseks. Selleks puhuks on välja arendatud [**CoNLL-U** formaat](https://universaldependencies.org/format.html), mis võimaldab puudepangad tekstifailidena talletada. Näide CoNLL-U formaadis lausest:

~~~
# sent_id = aja_luup200009_298
# text = Kahjuks saavad häid otsuseid teha vaid asjatundlikud isikud.
1	Kahjuks	kahjuks	ADV	D	_	5	advmod	5:advmod	_
2	saavad	saama	AUX	V	Mood=Ind|Number=Plur|Person=3|Tense=Pres|VerbForm=Fin|Voice=Act	5	aux	5:aux	_
3	häid	hea	ADJ	A	Case=Par|Degree=Pos|Number=Plur	4	amod	4:amod	_
4	otsuseid	otsus	NOUN	S	Case=Par|Number=Plur	5	obj	5:obj	_
5	teha	tegema	VERB	V	VerbForm=Inf	0	root	0:root	_
6	vaid	vaid	ADV	D	_	8	advmod	8:advmod	_
7	asjatundlikud	asja_tundlik	ADJ	A	Case=Nom|Degree=Pos|Number=Plur	8	amod	8:amod	_
8	isikud	isik	NOUN	S	Case=Nom|Number=Plur	5	nsubj	5:nsubj	SpaceAfter=No
9	.	.	PUNCT	Z	_	5	punct	5:punct	_
~~~

CoNLL-U formaadis on laused eraldatud tühjade ridadega. Lause sõnad on eraldi ridadel, mis algavad sõna ID-ga ja sisaldavad *tab*-iga eraldatud kindlas järjestuses tunnuseid:

| Väli          | seletus               | näide         |
| :------------ |:--------------------- |:-------       |
| ID            | järjekorranumber      | 2             |
| FORM          | sõnavorm              | saavad        |
| LEMMA         | lemma                 | saama         |
| UPOS          | universaalne sõnaliik | AUX           |
| XPOS          |keelespetsiifiline sõnaliik| V         |
| FEATS         | morfoloogilised tunnused | Mood=Ind\|Number=Plur\|Person=3 |
| HEAD          | sõna ülemuse id (või 0)| 5            |
| DEPREL        | sõltuvussüntaktiline märgend | aux    |
| DEPS          | täiendavad sõltuvusmärgendused* | 5:aux|
| MISC          | muud märgendused      | SpaceAfter=No |

\* Kuna üks ülemus ja sõltuvus ei pruugi anda edasi kõiki sõnadevahelisi seoseid (mõni seos võib olla ka rangete reeglitega vastuolus), on nende jaoks loodud täiendavate sõltuvusmärgenduste lahter.


Puuduvad tunnused on asendatud alakriipsuga, Kommentaarid ja metaandmed (lausenumber, failinimi jms) on välja toodud märgiga # algavatel ridadel.

## XML-andmeformaat

XML andmeformaadi eesmärgiks pakkuda andmete salvestamise ja levitamise viisi, mis oleks ühtviisi loetav nii inimesele kui ka arvutile. Ajalooliselt on XML olnud standardiks juba kaua ning seetõttu on salvestatud paljud varasemad tekstikogud ja korpused sellesse formaati. XML-i "sugulaskeel" on HTML (mõlemad pärinevad ühisest esivanemast: [SGML standardist](https://et.wikipedia.org/wiki/%C3%9Chtlustatud_%C3%9Cldine_Markeerimise_Keel)), mis on sisuliselt veebilehekülgede "lähtekoodi" keel – seega on enamik veebis olevast (pool)tekstilisest materjalist salvestatud just selles formaadis.

### _BeautifulSoup_

Alustuseks vaatame viise, kuidas XML-i *parsida* ehk automaatselt analüüsida XML-struktuuri ja otsida sealt mingeid alamosi. Pythonis on üheks väga heaks ja sageli kasutatavaks lahenduseks teek [_BeautifulSoup_](https://www.crummy.com/software/BeautifulSoup/). 

_\*Parsimise all mõistetakse süntaktilise analüüsi teostamist nii loomuliku keele kui ka programmeerimis- või märgenduskeele puhul. MaltParser ja VISLCG tegelevad arusaadavalt loomuliku keele parsimisega, beautifulsoup aga märgenduskeeltega. Seega, XML-ist ja HTML-ist rääkides mõistame parsimise all seda, kuidas märgendite vahelt oma tekst välja korjata._

Kuna _BeautifulSoup_ (moodul `bs4`) on üheks `estnltk` sõltuvuseks, peaks see teil `conda` keskkonda olema juba varasemalt installitud. Kui moodul `bs4` on mingil veidral põhjusel siiski puudu, tuleks see käsurealt installida:

    conda install bs4

Sama kehtib teegi `lxml` puhul. Seda saab vajadusel installida käsuga:

    conda install lxml    

Järgnevas näites on XML-kujul lõik ühest märgendatud ilukirjandustekstist, *BeautifulSoup*i abil parsime selle struktuuri ja kuvame "ilustatud" kujul, nii et taanded toovad XML-i struktuuri esile:

```python
from bs4 import BeautifulSoup

# Näite-XML ( lõigud pärinevad Tasakaalus korpuse ilukirjanduse failist 'tanav_jutt4.tei' )
xml_content = '''
<?xml version="1.0"?>
<text> <body> <div0 type='tervikteos'> <head> Taadeldus </head>
<p> <bibl> <author> <s> Maniakkide T&auml;nav </s> </author> </bibl> </p> <p> /---/ </p>
<p> <s> " L&auml;hme Widgeti juurde , " &uuml;tlesin . </s> <s> Widget oli kursa&otilde;de , kes elas siinsamas Annelinnas . </s> <s> Tee peal &uuml;ritas Ults mulle asja selgitada . </s> </p> <p> /---/ </p>
</div0> </body> </text>
'''

# Parsime XML-sisu lxml parseri abil
soup = BeautifulSoup(xml_content,'lxml')

# Kuvame XML elementide sisud ilusti joondatult:
print(soup.prettify())
```

* **Märkus HTML olemite kohta:** Varasemates XMLi versioonides polnud kodeeringu _'utf-8'_ kasutamine otseselt toetatud ning seetõttu tuli tekstide salvestamisel _'utf-8'_ sümbolid, sealhulgas ka eesti täpitähed, konverteerida nn *HTML-olemiteks*. Seetõttu võimegi eelmises näites täheldada olemeid `&uuml;` (tähistab tähte *ü*) ja `&otilde;` (tähistab tähte *õ*). Kui kasutada *BeautifulSoup*'i XML (või HTML) failide parsimiseks, ei pea olemite konverteerimise pärast muretsema: teek konverteerib olemid automaatselt, nagu on näha ka "ilustatud" väljundis.

### Lisamaterjal 2: EstNLTK funktsioonid koondkorpuse XML failide lugemiseks 
### (NB! Allpool järgneb veel HTML-andmeformaadi tutvustus ja üks ülesanne!)

Vaatasime juba, kuidas saab Eesti keele UD puudepanka Text-objektide kujule saada.
EstNLTKs on olemas vahendid veel kolme erineva korpuse sisselugemiseks: [Eesti keele koondkorpus](https://www.cl.ut.ee/korpused/segakorpus/), [eesti keele veebikorpus 2013](https://metashare.ut.ee/repository/browse/veebikorpus13-ettenten-toortekst/b564ca760de111e6a6e4005056b4002419cacec839ad4b7a93c3f7c45a97c55f/) (etTenTen) ja [Eesti keele ühendkorpus 2017](https://metashare.ut.ee/repository/browse/eesti-keele-uhendkorpus-2017/b616ceda30ce11e8a6e4005056b40024880158b577154c01bd3d3fcfc9b762b3/). Viimase alakorpuste hulgas on ka koondkorpus ja veebikorpus 2013. Juhendeid nende importimiseks leiab [siit](https://nbviewer.jupyter.org/github/estnltk/estnltk/blob/version_1.6/tutorials/corpus_processing/importing_text_objects_from_corpora.ipynb).

Vaatame siinkohal näitena, kuidas lugeda sisse Tartu Ülikooli koondkorpuse XML-formaadis tekste. Selleks tuleb esmalt koondkorpuse lehelt fail(id) alla laadida (linke järgides jõuab sealt koondkorpuse kokkupakitud XML-failideni). 

Tekstide sisselugemiseks on EstNLTKs olemas eraldiseisvad funktsioonid, koondkorpuse puhul leiab need moodulist `estnltk.corpus_processing.parse_koondkorpus`:

  * `parse_tei_corpus(path, target=['artikkel'])` – loeb ja parsib ühe XML faili (antud kataloogiteega `path`) sisu ning tagastab faili põhjal loodud Text-objektide järjendi; 


  * `parse_tei_corpora(root, prefix='', suffix='.xml', target=['artikkel'])` – loeb ja parsib kataloogist `root` faile, mille prefiks on `prefix` ja sufiks `suffix` (juba vaikimisi `prefix=''`, `suffix='.xml'`) ning tagastab failide põhjal loodud Text-objektide järjendi.
  
Mõlema eelmise funktsiooni puhul otsitakse tekstilist sisu `div`-elementidest, mille `type`-väärtused on kirjeldatud järjendis `target`. Vaikeväärtus `target=['artikkel']` sobib enamiku ajakirjandustekstide parsimiseks, ilukirjanduse puhul peaks sobima väärtus `target=['tervikteos']`. Kui aga pole teada, milline väärtus on õige (nt foorumitekstide või jututoatekstide puhul), tagastab järgmine funktsioon vastavale tekstile loogilise alamjaotuse väärtuse:
 
  * `get_div_target(input_xml_filepath)`
  
**NB!** Meetodi kasutamisel peab sisendiks olema täistee. Kui säilitatud pole algne allalaadimisel tekkinud kaustatee (nt ühe foorumiteksti puhul *...\jututoad\xml\failid\blue121103.xml*), ei pruugi meetod anda korrektset tulemust ja vaikimisi tagastatakse **'artikkel'**.

#### Lausestuse säilitamisest

Mooduli `estnltk.corpus_processing.parse_koondkorpus` funktsioonide abil vaikimisi sätetega tekste sisse lugedes ei lisata märgenduskihte. Lisades korpust parsides lipu `add_tokenization=True` määratakse, et Text-objektid saavad kohe ka esialgse märgenduse (samad kihid, mis tag_layer() rakendamisel). Sellisel juhul võite aga avastada, et tekstide lausestus ei tule päris selline, nagu see oli algses XML-failis `<s>`-märgendite abil esile toodud.

Kuna koondkorpuse XML-failis puudub päris algne (ilma märgendusteta) tekst, peavad meetodid `parse_tei_corpora()` ja `parse_tei_corpus()` teksti rekonstrueerima. 
Rekonstrueerimise käigus paigutatakse iga `<s>`-märgendite vaheline tekstisisu vaikimisi eraldi reale (ehk siis: laused on eraldatud reavahetustega). 
EstNLTK tavapärane lausestaja aga sellist lausestust ei tunne, vaid otsib lausepiire ikka punktuatsioonisümbolite järgi.
Mõnikord EstNLTK tavalausestaja eksib ning arvab, et XML-failis lause keskel olev punktuatsioon on lausepiiriks, ning seetõttu tekivadki sisseloetud tekste lausestades teistsugused lausepiirid, kui olid algses XML-failis.* 

Selleks, et saada täpselt samasugune lausestus nagu oli algses XML-failis, tuleb korpuse parsimisel ette anda lipp `preserve_tokenization=True`, see aga vajab, et sisse oleks lülitatud ka `add_tokenization`.

\* Tuleb ära märkida, et korpuse lausestuse säilitamine ei pruugi samuti anda täiuslikku tulemust – ka korpuste märgendamiseks on kasutatud automaatlausestajat, mis tahes-tahtmata on vigu teinud.
