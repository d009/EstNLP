### Süntaktiline analüüs

Süntaktilise analüüsi (sünonüümina kasutatakse sõna "parsima") ülesandeks on lause struktuuri ehk sõnadevaheliste seoste leidmine. Kui me teostame tekstil morfoloogilise analüüsi ja leiame sõnaliigid, siis saame üldjoontes teada, milliseid tegevusi selle lausega kirjeldatakse (tegusõnad), millistest olenditest/objektidest/nähtustest jne seal räägitakse (nimisõnad), milliseid omadusi seal mainitakse (omadussõnad) jne. Sõnaliigid ei anna aga meile infot sõnadevahelistest seostest. Näiteks, kui teame, et lauses esinevad nimisõnad "ema" ja "isa" ning tegusõna "hirmutama", siis ei ole meil siiski teadmist selle kohta, kes keda ehmatab. Kui teame aga, et "ema" on lauses alus (subjekt), "isa" on sihitis (objekt) ning "hirmutama" on selle lause öeldis, siis on selge, et ema on hirmutajaks ja isa selleks, keda hirmutatakse. Seda teadmist saame esitada puu kujul:

Selles praktikumis vaatame kaht süntaktiliseks märgendamiseks kasutatavat töövahendit: [MaltParseri](http://www.maltparser.org/) ja [Stanza](https://stanfordnlp.github.io/stanza/) analüsaatoreid. Mõlemat on võimalik kasutada läbi EstNLTK. Siin kasutatavaid **süntaksimärgendeid aitab mõista [see](https://universaldependencies.org/u/dep/index.html) leht**.

### MaltParser

Esmalt katsetame süntaktilise analüüsi jaoks MaltParserit. MaltParseri kasutamisel on võimalik väljastada mitme märgendussüsteemi märgendeid. Selles praktikumis keskendume [Universal Dependencies](https://universaldependencies.org/#-universal-dependencies) raamistiku (edaspidi UD) märgendussüsteemile, sellest saab rohkem lugeda sama praktikumi lisamaterjali all.

MaltParser nõuab [Java](https://www.java.com/en/download/) olemasolu süsteemis. 
Kõige lihtsam viis MaltParseri kasutamiseks on läbi vaiketööahela -- kutsudes tekstil välja `tag_layer('maltparser_syntax')`:

```python
from estnltk import Text

text = Text('Poe ees õlut joonud mehed häirisid kohalikke, aga ükskõikne politsei ei teinud midagi.')

text.tag_layer('maltparser_syntax')
```

MaltParseri töötluse käigus lisatakse tekstile 2 kihti:
* `maltparser_conll_morph` -- MaltParseri sisendkiht, mis sisaldab morfoloogilist analüüsi teisendatuna sellisele kujule, mille  pealt MaltParser oskab ennustada süntaksiseoseid;
* `maltparser_syntax` -- MaltParseri väljundkiht, kus on märgitud sõnadevahelised sõltuvusseosed (`id`, `head`, `parent_span`, `children`) ja sõnade süntaktilised märgendid (`deprel`);

```python
# Uurime MaltParseri väljundkihti
text.maltparser_syntax
```

Atribuut `deprel` sisaldab parseri lisatud süntaktilist märgendit, nt _nsubj_ märgib (osa)lause subjekti, _obj_ objekti, _root_ lause juurt (üldjuhul finiitne põhiverb, öeldis). Keerukamad on osalausete vahelised suhted. Nagu näeme, saab teise osalause finiitne verb "teinud" külge märgendi _conj_, mis viitab sellele, et osalaused on rinnastatud ehk seotud rinnastavate sidesõnadega. **Tutvu kõigi märgenditega [sellel](https://universaldependencies.org/u/dep/index.html) lehel**.

Atribuut `head` viitab lauseliikme ülemuse *id*-le. Näiteks allub subjekt "mehed" lauseliikmele, mille id on 6, ehk öeldisele "häirisid". Viimane omakorda on lause **juur**, st ei allu ühelegi teisele lauseliikmele, sellele viitab *head*-atribuudi väärtus 0.

MaltParseri tagastatav analüüs on ühene.

#### Puustruktuuri uurimine
Süntaksis ei huvitu me enamasti vaid ühest sõnast korraga, vaid ka sõnadevahelistest seostest. Nagu võib aimata varem läbi käinud mõistetest *juur* ja *ülemus*, moodustavad sõnad ja nendevahelised suhted (süntaksi)puu. Selleks, et ülemustele-alluvatele lihtsamini ligi pääseda ja meid huvitavaid konstruktsioone lihtsamini läbi käia, saame vaikimisi kätte ka atribuudid `parent_span` ja `children` (viitavad ülemuste-alluvate Span'idele). Vaatame näidet kahelauselise teksti puhul:

```python
text = Text("Pisike poiss kimbutas kirpe. Need kirbud ei julgenud kiusata karusid. Koer ja kass sõid putru.")
text.tag_layer('maltparser_syntax')
display(text.maltparser_syntax)
```

Itereerime üle lausete finiitverbide alluvate ja väljastame nende ülemuse ja süntaktilise märgendi:

```python
for word in text.maltparser_syntax:
    if word.deprel == 'root':
        for child in word.children:
            print(word.text, '-->', child.deprel, child.text)
```

#### MaltParserTagger-i kasutamine

Kui soovid enne märgendamist muuta MaltParseri konfiguratsiooni (või mudelit), siis peaksid looma uue MaltParserTagger-i ning teostama teksti märgendamise selle abil.

```python
from estnltk.taggers import MaltParserTagger
maltparser_tagger = MaltParserTagger(version='conllu') # määrame parameetri version väärtuseks 'conllu', et saada UD märgendeid
maltparser_tagger
```

Märgendused lisatakse kihti *maltparser_syntax*, aga nagu näha, peavad märgendamiseks olema tekstil kihid `words`, `sentences` ja `conll_morph`. Viimase kihi märgendamiseks peame kasutama `ConllMorphTaggerit`. UD märgendust soovides on oluline seada lipp `no_visl=True` (vastasel juhul kasutab Tagger sisemiselt morfoloogilise analüüsi kihi teisendamiseks [VISLCG3 analüsaatorit](https://github.com/EstSyntax/EstCG), aga selle kasutamise eeldused meil puuduvad ning korrektse UD väljundi saamiseks seda ka ei kasutata). Kuna märgendame oma teksti `morph_analysis` kihi põhjal, tuleb kihi nimi seada ka parameetri `morph_extended_layer` väärtuseks. 

```python
from estnltk.taggers import ConllMorphTagger
conllmorph_tagger = ConllMorphTagger(morph_extended_layer='morph_analysis', no_visl=True)

# Vaatame, milliseid märgenduskihte vajab ConllMorphTagger
conllmorph_tagger.input_layers
```

Eeltöötleme nüüd teksti MaltParseri kasutamiseks: vaja on kihte *words*, *sentences*, *morph_analysis* ja *conll_morph*:

```python
from estnltk import Text

# Lisame default_layers kihid
text = Text('Ilus suur karvane kass nurrus punasel diivanil').tag_layer('morph_analysis')

# Lisame conll_morph kihi
conllmorph_tagger.tag(text)
```

```python
# Lõpuks lisame MaltParseri kihi
maltparser_tagger.tag(text)
text.maltparser_syntax
```

**Lisaks:** EstNLTKs saab MaltParserTaggerit kasutada ka alternatiivsete mudelitega, mida me selles praktikumis ei käsitle. Näiteks on võimalik kasutada mudelit, mis töötab morph_extended kihi baasil või mudeleid, mille väljund on alternatiivses (kitsenduste grammatika) märgendisüsteemis. Need mudelid tuleb aga kasutamiseks ise alla laadida ning ConllMorphTaggeri ja MaltParserTaggerit vastavalt vajadusele seadistada. Täpsema info mudelite kohta ning juhendid Taggerite seadistamiseks leiab [siit](https://github.com/estnltk/estnltk/blob/main/tutorials/nlp_pipeline/C_syntax/03_syntactic_analysis_with_maltparser.ipynb) (vt alamosa MaltParserTagger).

### StanzaSyntaxWebTagger

Peale Maltparseri saame süntaktilisi märgendeid lisada ka [Stanza](https://stanfordnlp.github.io/stanza/) süntaksianalüsaatori abil.
Selleks impordime `StanzaSyntaxWebTagger`'i, mis märgendab etteantud teksti sõltuvussüntaktiliselt EstNLTK StanzaSyntaxTaggeri veebiteenuse kaudu. Kindlasti tuleb taggeri initsialiseerimisel ette anda vastav url.

```python
from estnltk.taggers import StanzaSyntaxWebTagger
stanza_syntax_web_tagger = \
    StanzaSyntaxWebTagger(url='https://api.tartunlp.ai/estnltk/tagger/stanza_syntax')
stanza_syntax_web_tagger
```

Nagu näha, on StanzaSyntaxWebTaggeri kasutamiseks vaja, et Text-objektil oleks lisaks sõnade ja lausete kihile olemas ka kiht `morph_extended`.  Kihi *morph_extended* ülemkiht on *morph_analysis* (kasutades `tag_layer('morph_extended')` lisatakse Text-objektile automaatselt ka *morph_analysis* kiht).

Lisame `morph_extended` kihi ja vaatame, mis moel see tavapärasest morfoloogilise analüüsisi kihist erineb.

```python
from estnltk import Text
text = Text('Poe ees õlut joonud mehed häirisid kohalikke, aga ükskõikne politsei ei teinud midagi. Mina läksin ära.')
```

```python
text.tag_layer('morph_extended')
display(text['morph_extended'])
```

Näeme, et võrreldes 'morph_analysis' kihiga sisaldab 'morph_extended' kihis atribuut *form* detailsemat infot ja juures on ka mitu uut atribuuti (nt *punctuation_type* ja *letter_case*).

```python
# Loome teksti ja märgendame nõutud kihid
from estnltk import Text
text = Text('Ilus suur karvane kass nurrus rohelisel diivanil.').tag_layer('morph_extended')

stanza_syntax_web_tagger.tag( text )
text.stanza_syntax
```

StanzaSyntaxWebTaggeriga ei saa me kätte `parent_span` ja `children` atribuute - kui siiski tahame, et need kihid lisaks oleksid, tuleb kasutada  `SyntaxDependencyRetagger`'it (rohkem informatsiooni [sellel lehel](https://github.com/estnltk/estnltk/blob/main/tutorials/nlp_pipeline/C_syntax/04_syntactic_analysis_utilities.ipynb)). Taggeri loomisel peame täpsustama süntaksikihi, millele *parent_span* ja *children* atribuute lisada soovime.

```python
from estnltk.taggers import SyntaxDependencyRetagger

SyntaxDependencyRetagger('stanza_syntax').retag(text)

text.stanza_syntax
```

**Lisaks:** Stanzat saab veel kasutada ka StanzaSyntaxEnsembleWebTagger-i kaudu. Selle eripära on süntaksimärgendi valimisel mitme Stanza süntaksimudeli põhjal otsuse tegemine. Seega võiks see olla täpsem kui ühel mudelil põhinev StanzaSyntaxWebTagger, kuid samas ajakulukam. Vaata StanzaSyntaxEnsembleWebTagger-i kasutamist [WebTaggerite juhendist](https://github.com/estnltk/estnltk/blob/main/tutorials/taggers/web_taggers/web_taggers.ipynb).

```python
from estnltk.taggers import SyntaxDependencyRetagger
# Tekitame teksti koos morph_extended kihiga
text = Text('Ilus suur karvane kass nurrus rohelisel diivanil.').tag_layer('morph_extended')
# Lisame MaltParseri süntaksikihi
text.tag_layer('maltparser_syntax')
# Lisame Stanza süntaksikihi
stanza_syntax_web_tagger.tag( text )
```

### Lisamaterjal: Universal Dependencies ja CoNLL-U formaat

[Universal Dependencies](https://universaldependencies.org/#language-) (UD) on projekt, mille eesmärk on luua eri keeltele sobiv morfoloogiline ja sõltuvussüntaktiline märgendussüsteem ning seda järgivad puudepangad, mis võimaldaksid nt mitmekeelse parseri arendamise ja keeltevahelise masinõppe. Seega püüab UD universaalsete kategooriate abil kindlustada keelest sõltumatult järjepidevat sarnaste konstruktsioonide kirjeldamist, samas lubades vajadusel teatavaid keelespetsiifilisi täiendusi. UD süntaksimärgendamise põhimõtetest saab rohkem lugeda [siit](https://universaldependencies.org/u/overview/syntax.html).

UD peab olema rahuldav paljude keelte (tüpoloogia) kirjeldamiseks, samas lihtsasti mõistetav nii inimesele (sh mitte-lingvistile) kui ka sobiv arvutiga parsimiseks ja töötlemiseks. Selleks puhuks on välja arendatud [**CoNLL-U** formaat](https://universaldependencies.org/format.html), mis võimaldab puudepangad tekstifailidena talletada. Näide CoNLL-U formaadis lausest:

~~~
# sent_id = aja_luup200009_298
# text = Kahjuks saavad häid otsuseid teha vaid asjatundlikud isikud.
1	Kahjuks	kahjuks	ADV	D	_	5	advmod	5:advmod	_
2	saavad	saama	AUX	V	Mood=Ind|Number=Plur|Person=3|Tense=Pres|VerbForm=Fin|Voice=Act	5	aux	5:aux	_
3	häid	hea	ADJ	A	Case=Par|Degree=Pos|Number=Plur	4	amod	4:amod	_
4	otsuseid	otsus	NOUN	S	Case=Par|Number=Plur	5	obj	5:obj	_
5	teha	tegema	VERB	V	VerbForm=Inf	0	root	0:root	_
6	vaid	vaid	ADV	D	_	8	advmod	8:advmod	_
7	asjatundlikud	asja_tundlik	ADJ	A	Case=Nom|Degree=Pos|Number=Plur	8	amod	8:amod	_
8	isikud	isik	NOUN	S	Case=Nom|Number=Plur	5	nsubj	5:nsubj	SpaceAfter=No
9	.	.	PUNCT	Z	_	5	punct	5:punct	_
~~~

CoNLL-U formaadis on laused eraldatud tühjade ridadega. Lause sõnad on eraldi ridadel, mis algavad sõna ID-ga ja sisaldavad *tab*-iga eraldatud kindlas järjestuses tunnuseid, sõnaliigi UPOS tähendused leiad [sellelt lehelt](https://universaldependencies.org/u/pos/index.html):

| Väli          | seletus               | näide         |
| :------------ |:--------------------- |:-------       |
| ID            | järjekorranumber      | 2             |
| FORM          | sõnavorm              | saavad        |
| LEMMA         | lemma                 | saama         |
| UPOS          | universaalne sõnaliik | AUX           |
| XPOS          |keelespetsiifiline sõnaliik| V         |
| FEATS         | morfoloogilised tunnused | Mood=Ind\|Number=Plur\|Person=3 |
| HEAD          | sõna ülemuse id (või 0)| 5            |
| DEPREL        | süntaktiline märgend | aux    |
| DEPS          | täiendavad süntaksimärgendused* | 5:aux|
| MISC          | muud märgendused      | SpaceAfter=No |

\* Kuna üks ülemus ja süntaksimärgend ei pruugi anda edasi kõiki sõnadevahelisi seoseid (mõni seos võib olla ka rangete reeglitega vastuolus), on nende jaoks loodud täiendavate märgenduste lahter.


Puuduvad tunnused on asendatud alakriipsuga, Kommentaarid ja metaandmed (lausenumber, failinimi jms) on välja toodud märgiga # algavatel ridadel.

## XML-andmeformaat

XML-andmeformaadi eesmärgiks pakkuda andmete salvestamise ja levitamise viisi, mis oleks ühtviisi loetav nii inimesele kui ka arvutile. Ajalooliselt on XML olnud standardiks juba kaua ning seetõttu on salvestatud paljud varasemad tekstikogud ja korpused sellesse formaati. XML-i "sugulaskeel" on HTML (mõlemad pärinevad ühisest esivanemast: [SGML standardist](https://et.wikipedia.org/wiki/%C3%9Chtlustatud_%C3%9Cldine_Markeerimise_Keel)), mis on sisuliselt veebilehekülgede "lähtekoodi" keel – seega on enamik veebis olevast (pool)tekstilisest materjalist salvestatud just selles formaadis.

### _BeautifulSoup_

Alustuseks vaatame viise, kuidas XML-i *parsida* ehk automaatselt analüüsida XML-struktuuri ja otsida sealt mingeid alamosi. Pythonis on üheks väga heaks ja sageli kasutatavaks lahenduseks teek [_BeautifulSoup_](https://www.crummy.com/software/BeautifulSoup/). 

_\*Parsimise all mõistetakse süntaktilise analüüsi teostamist nii loomuliku keele kui ka programmeerimis- või märgenduskeele puhul. MaltParser ja Stanza tegelevad arusaadavalt loomuliku keele parsimisega, beautifulsoup aga märgenduskeeltega. Seega, XML-ist ja HTML-ist rääkides mõistame parsimise all seda, kuidas märgendite vahelt oma tekst välja korjata._

Kuna _BeautifulSoup_ (moodul `bs4`) on üheks `estnltk` sõltuvuseks, peaks see teil `conda` keskkonda olema juba varasemalt installitud. Kui moodul `bs4` on mingil veidral põhjusel siiski puudu, tuleks see käsurealt installida:

    conda install bs4

Sama kehtib teegi `lxml` puhul. Seda saab vajadusel installida käsuga:

    conda install lxml    

Järgnevas näites on XML-kujul lõik ühest märgendatud ilukirjandustekstist, *BeautifulSoup*i abil parsime selle struktuuri ja kuvame "ilustatud" kujul, nii et taanded toovad XML-i struktuuri esile:

```python
from bs4 import BeautifulSoup

# Näite-XML ( lõigud pärinevad Tasakaalus korpuse ilukirjanduse failist 'tanav_jutt4.tei' )
xml_content = '''
<?xml version="1.0"?>
<text> <body> <div0 type='tervikteos'> <head> Taadeldus </head>
<p> <bibl> <author> <s> Maniakkide T&auml;nav </s> </author> </bibl> </p> <p> /---/ </p>
<p> <s> " L&auml;hme Widgeti juurde , " &uuml;tlesin . </s> <s> Widget oli kursa&otilde;de , kes elas siinsamas Annelinnas . </s> <s> Tee peal &uuml;ritas Ults mulle asja selgitada . </s> </p> <p> /---/ </p>
</div0> </body> </text>
'''

# Parsime XML-sisu lxml parseri abil
soup = BeautifulSoup(xml_content,'lxml')

# Kuvame XML elementide sisud ilusti joondatult:
print(soup.prettify())
```

* **Märkus HTML olemite kohta:** Varasemates XMLi versioonides polnud kodeeringu _'utf-8'_ kasutamine otseselt toetatud ning seetõttu tuli tekstide salvestamisel _'utf-8'_ sümbolid, sealhulgas ka eesti täpitähed, konverteerida nn *HTML-olemiteks*. Seetõttu võimegi eelmises näites täheldada olemeid `&uuml;` (tähistab tähte *ü*) ja `&otilde;` (tähistab tähte *õ*). Kui kasutada *BeautifulSoup*'i XML (või HTML) failide parsimiseks, ei pea olemite konverteerimise pärast muretsema: teek konverteerib olemid automaatselt, nagu on näha ka "ilustatud" väljundis.

### Lisamaterjal 2: EstNLTK funktsioonid koondkorpuse XML failide lugemiseks 
### (NB! Allpool järgneb veel HTML-andmeformaadi tutvustus ja üks ülesanne!)

Vaatasime juba, kuidas saab Eesti keele UD puudepanka Text-objektide kujule saada.
EstNLTKs on olemas vahendid veel kolme erineva korpuse sisselugemiseks: [Eesti keele koondkorpus](https://www.cl.ut.ee/korpused/segakorpus/), [Eesti keele veebikorpus 2013](https://metashare.ut.ee/repository/browse/veebikorpus13-ettenten-toortekst/b564ca760de111e6a6e4005056b4002419cacec839ad4b7a93c3f7c45a97c55f/) (etTenTen) ja [Eesti keele ühendkorpus 2021](https://doi.org/10.15155/3-00-0000-0000-0000-08E60L). Viimase alakorpuste hulgas on ka koondkorpus ja veebikorpus 2013. Juhendeid nende importimiseks leiab [siit](https://nbviewer.org/github/estnltk/estnltk/blob/main/tutorials/corpus_processing/importing_text_objects_from_corpora.ipynb).

Vaatame siinkohal näitena, kuidas lugeda sisse Tartu Ülikooli koondkorpuse XML-formaadis tekste. Selleks tuleb esmalt koondkorpuse lehelt fail(id) alla laadida (linke järgides jõuab sealt koondkorpuse kokkupakitud XML-failideni). 

Tekstide sisselugemiseks on EstNLTKs olemas eraldiseisvad funktsioonid, koondkorpuse puhul leiab need moodulist `estnltk.corpus_processing.parse_koondkorpus`:

  * `parse_tei_corpus(path, target=['artikkel'])` – loeb ja parsib ühe XML faili (antud kataloogiteega `path`) sisu ning tagastab faili põhjal loodud Text-objektide järjendi; 


  * `parse_tei_corpora(root, prefix='', suffix='.xml', target=['artikkel'])` – loeb ja parsib kataloogist `root` faile, mille prefiks on `prefix` ja sufiks `suffix` (juba vaikimisi `prefix=''`, `suffix='.xml'`) ning tagastab failide põhjal loodud Text-objektide järjendi.
  
Mõlema eelmise funktsiooni puhul otsitakse tekstilist sisu `div`-elementidest, mille `type`-väärtused on kirjeldatud järjendis `target`. Vaikeväärtus `target=['artikkel']` sobib enamiku ajakirjandustekstide parsimiseks, ilukirjanduse puhul peaks sobima väärtus `target=['tervikteos']`. Kui aga pole teada, milline väärtus on õige (nt foorumitekstide või jututoatekstide puhul), tagastab järgmine funktsioon vastavale tekstile loogilise alamjaotuse väärtuse:
 
  * `get_div_target(input_xml_filepath)`
  
**NB!** Meetodi kasutamisel peab sisendiks olema täistee. Kui säilitatud pole algne allalaadimisel tekkinud kaustatee (nt ühe foorumiteksti puhul *...\jututoad\xml\failid\blue121103.xml*), ei pruugi meetod anda korrektset tulemust ja vaikimisi tagastatakse **'artikkel'**.

#### Lausestuse säilitamisest

Mooduli `estnltk.corpus_processing.parse_koondkorpus` funktsioonide abil vaikimisi sätetega tekste sisse lugedes ei lisata märgenduskihte. Lisades korpust parsides lipu `add_tokenization=True` määratakse, et Text-objektid saavad kohe ka esialgse märgenduse (samad kihid, mis tag_layer() rakendamisel). Sellisel juhul võite aga avastada, et tekstide lausestus ei tule päris selline, nagu see oli algses XML-failis `<s>`-märgendite abil esile toodud.

Kuna koondkorpuse XML-failis puudub päris algne (ilma märgendusteta) tekst, peavad meetodid `parse_tei_corpora()` ja `parse_tei_corpus()` teksti rekonstrueerima. 
Rekonstrueerimise käigus paigutatakse iga `<s>`-märgendite vaheline tekstisisu vaikimisi eraldi reale (ehk siis: laused on eraldatud reavahetustega). 
EstNLTK tavapärane lausestaja aga sellist lausestust ei tunne, vaid otsib lausepiire ikka punktuatsioonisümbolite järgi.
Mõnikord EstNLTK tavalausestaja eksib ning arvab, et XML-failis lause keskel olev punktuatsioon on lausepiiriks, ning seetõttu tekivadki sisseloetud tekste lausestades teistsugused lausepiirid, kui olid algses XML-failis.* 

Selleks, et saada täpselt samasugune lausestus nagu oli algses XML-failis, tuleb korpuse parsimisel ette anda lipp `preserve_tokenization=True`, see aga vajab, et sisse oleks lülitatud ka `add_tokenization`.

\* Tuleb ära märkida, et korpuse lausestuse säilitamine ei pruugi samuti anda täiuslikku tulemust – ka korpuste märgendamiseks on kasutatud automaatlausestajat, mis tahes-tahtmata on vigu teinud.

```python
from bs4 import BeautifulSoup
```

```python
html = '''<!DOCTYPE html>
<html>
<body>

<h1>Pole probleemi</h1>

<p>Oli õhtune tipptund, tuli sõita läbi linna.</p>
<p>Tuled fooris vahetusid, kuid üle veel ei saanud.</p>

</body>
</html>'''
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
for i in soup.find_all('p')[:10]:
    print(i)
    print(i.get_text())
```
